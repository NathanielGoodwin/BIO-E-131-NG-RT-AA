```python
import sys
infile = open("romeoandjuliet.txt")
for line in infile:
    words = line.split()
    for word in words:
        word = word.lower()
        for punc in '-.,;()[]?!\n':
                word=word.replace(punc,'')
        if word in newdict:
            newdict[word]=newdict[word]+1
        else: 
            newdict[word]=1
print(newdict)
            
```
