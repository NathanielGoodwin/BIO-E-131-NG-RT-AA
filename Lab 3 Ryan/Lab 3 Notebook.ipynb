Adding these words to dict w/ count:

```python
import string
translator = str.maketrans(string.ascii_uppercase, string.ascii_lowercase, string.punctuation)
dict = {}
with open("romeoandjuliet.txt") as file:
  for line in file:
    for word in line.split():
      word = word.translate(translator)
      if word in dict:
        dict[word] += 1
      else:
        dict[word] = 1
print(dict)
```

creating table using plotly:

```python
import matplotlib.pyplot as plt
import string
translator = str.maketrans(string.ascii_uppercase, string.ascii_lowercase, string.punctuation)
dict = {}
with open("romeoandjuliet.txt") as file:
  for line in file:
    for word in line.split():
      word = word.translate(translator)
      if word in dict:
        dict[word] += 1
      else:
        dict[word] = 1
        
plt.bar(range(len(dict)), list(dict.values()), align='center')
plt.xticks(range(len(dict)), list(dict.keys()))

plt.show()
```
