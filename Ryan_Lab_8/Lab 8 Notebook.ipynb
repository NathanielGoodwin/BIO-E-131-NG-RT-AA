Q) Why do we expect short reads to produce a more fragmented assembly than long reads?
A) The location of short reads is harder to determine than longer reads due to repetition in genomes and simply having less base pairs to compare with the genome around it. This results in a more fragmented assembly.

Q) Why does a single-molecule sequencing like PacBio have a higher error rate than Illumina?
A) As the read lengths increase, typically accuracy decreases, therefore genomes assembled from many short accurate strands have as a whole less errors.
